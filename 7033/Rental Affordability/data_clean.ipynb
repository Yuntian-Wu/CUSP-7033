```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('DOF_Condominium_Comparable_Rental_Income_in_NYC_20240504.csv')
```

```python
data.columns
```

```python
columns_to_drop = ['Condo Section', 'Distance from Condo in miles', 'Distance from Condo in miles 1', 'Distance from Condo in miles 2']
```

```python
columns_to_drop = [col for col in columns_to_drop if col in data.columns]
data.drop(columns_to_drop, axis=1, inplace=True)
```

```python
base_columns = [
    'Boro-Block-Lot', 'Address', 'Neighborhood', 'Building Classification',
    'Total Units', 'Year Built', 'Gross SqFt', 'Estimated Gross Income', 'Gross Income per SqFt',
    'Estimated Expense', 'Expense per SqFt', 'Net Operating Income', 'Full Market Value',
    'Market Value per SqFt'
]
```

Identify Columns with Indexes: Determine which columns have multiple indexed versions (1, 2, 3).

Stack the DataFrames: Combine the modified DataFrames into a single DataFrame.

```python
report_year = data['Report Year'].copy()
```

```python
dfs = []
for i in range(3):  
    col_suffix = '' if i == 0 else f' {i+1}'
    columns = [col + col_suffix if f'{col} {i+1}' in data.columns else col for col in base_columns]
    subset = data[columns]
    subset = subset.copy() 
    subset.columns = base_columns  # Standardize column names to facilitate combining
    subset.loc[:, 'Report Year'] = report_year 
    dfs.append(subset)
```

```python
combined_df = pd.concat(dfs, ignore_index=True)
```
